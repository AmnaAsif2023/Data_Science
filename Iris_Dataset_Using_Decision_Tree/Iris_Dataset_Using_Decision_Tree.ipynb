```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#independent feature and dependent features
x = df.iloc[:,:-1]
y = iris.target
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(x_train,y_train)
```

```python
y_pred=treemodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc = accuracy_score(y_pred,y_test)
```

```python
acc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.heatmap(cm,annot=True)
```

```python
y_pred = treemodel.predict(x_test)
conf_matrix = confusion_matrix(y_test, y_pred)
df_conf_matrix = pd.DataFrame(conf_matrix, 
                              columns=['Predicted Class 0', 'Predicted Class 1', 'Predicted Class 2'], 
                              index=['Actual Class 0', 'Actual Class 1', 'Actual Class 2'])

df_conf_matrix
```

```python
sns.heatmap(df_conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
